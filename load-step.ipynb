### Below block of code creates a staging scheme and the staging tables and inserts the data from delta file locations.

```
# Create the 'staging' schema. Its like a database holding the tables.
spark.sql("CREATE SCHEMA IF NOT EXISTS staging")
```

```
# Drop if the staging.riders table exists, Creates the riders table using delta file locations
spark.sql("DROP TABLE IF EXISTS staging.riders")
spark.sql("CREATE TABLE staging.riders USING DELTA LOCATION '/delta/data/riders'")

# Drop if the staging.payments table exists, Creates the riders table using delta file locations
spark.sql("DROP TABLE IF EXISTS staging.payments")
spark.sql("CREATE TABLE staging.payments USING DELTA LOCATION '/delta/data/payments'")

# Drop if the staging.stations table exists, Creates the riders table using delta file locations
spark.sql("DROP TABLE IF EXISTS staging.stations")
spark.sql("CREATE TABLE staging.stations USING DELTA LOCATION '/delta/data/stations'")

# Drop if the staging.trips table exists, Creates the riders table using delta file locations
spark.sql("DROP TABLE IF EXISTS staging.trips")
spark.sql("CREATE TABLE staging.trips USING DELTA LOCATION '/delta/data/trips'")
```
